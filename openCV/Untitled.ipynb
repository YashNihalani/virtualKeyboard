```python
import cv2
import numpy as np
import mediapipe as mp
from cvzone.HandTrackingModule import HandDetector
from time import sleep
from pynput.keyboard import Controller
```

```python
web_cam=cv2.VideoCapture(0)
web_cam.set(3,1280) 
# set resoltuion ?
web_cam.set(4,720)
```

```python
detector=HandDetector(detectionCon=0.8, maxHands=1)
# detection confidence is thing to learn
keys = [["Q", "W", "E", "R", "T", "Y", "U", "I", "O", "P"],
        ["A", "S", "D", "F", "G", "H", "J", "K", "L", ";"],
        ["Z", "X", "C", "V", "B", "N", "M", ",", ".", "/"]]
```

```python
def drawAll(img,buttonList):
    for button in buttonList:
        x,y=button.pos
        w,h=button.size
        cv2.rectangle(img,button.pos,(x+w,y+h),(255,0,255),cv2.FILLED)
        cv2.putText(img,button.text,(x+20,y+65),cv2.FONT_HERSHEY_PLAIN,5,(255,255,255),5) 
    return img

class Button():
    def __init__(self,pos,text,size=[90,80]):
        self.pos=pos
        self.size=size
        self.text=text

   
```

```python

buttonList=[]
for i in range (len(keys)):
    for j,key in enumerate (keys[i]):
        buttonList.append(Button([(100*j+100),((100*i)+50)],key))
        
# print (len(buttonList))
txt=""
keyboard=Controller()
while(1):
    success,img=web_cam.read()
    hands= detector.findHands(img,draw=False)
    img=drawAll(img,buttonList)
    if hands:
         hand1=hands[0]
         lmList=hand1['lmList']
         if lmList:
                for button in buttonList:
                    x,y=button.pos
                    w,h=button.size
                    
                    if x<=lmList[8][0]<=x+w and y<=lmList[8][1]<=y+h:
                        cv2.rectangle(img,button.pos,(x+w,y+h),(175,0,175),cv2.FILLED)
                        cv2.putText(img,button.text,(x+20,y+65),cv2.FONT_HERSHEY_PLAIN,5,(255,255,255),5)
                        l,info=detector.findDistance(lmList[8],lmList[12])
                        if l<30:
                            cv2.rectangle(img,button.pos,(x+w,y+h),(0,255,0),cv2.FILLED)
                            cv2.putText(img,button.text,(x+20,y+65),cv2.FONT_HERSHEY_PLAIN,5,(255,255,255),5)
#                             txt+=button.text
                            xx=button.text[0]
                            print(xx)
                            sleep(0.08)

                         
    cv2.imshow("IMAGE",img)
    if cv2.waitKey(1) & 0xff==27:
        break
cv2.destroyAllWindows() 
web_cam.release()
# just boiler plate code accessing video and reading it image by image 
    
```
